```python
avg_total = df.groupby("time")["total_bill"].mean()
plt.plot(avg_total.index,avg_total.values,marker=".",linestyle="-")
plt.title("Variation of sales over time")
plt.xlabel("Time")
plt.ylabel("total_bill")
plt.show()
```
