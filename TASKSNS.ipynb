```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Customer_support_data.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
missing = df.isnull().sum()
```

```python
df
```

```python
plt.figure(figsize=(12, 4))
missing[missing > 0].sort_values(ascending=False).plot(kind='bar', color='salmon')
plt.title("Missing Values per Column")
plt.ylabel("Number of Missing Values")
plt.show()
```

missing

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols].hist(figsize=(15, 10), bins=20, color='skyblue')
plt.suptitle("Distribution of Numeric Columns")
plt.show()
```

```python

sns.histplot(df["Item_price"],kde = True)
plt.title("Histogram of Item price")
```

```python
sns.boxplot(y = "CSAT Score",data = df)
plt.title('Box Plot - CSAT Score')
plt.grid()
plt.show()
```

```python
sns.violinplot(y="connected_handling_time",data = df,color = "skyblue")
plt.title("Distribution of connected handling time")
plt.grid(True)
plt.show()
```

```python
csat_counts = df["CSAT Score"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(csat_counts, labels=csat_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue','lightgreen','salmon','orange','violet'])
plt.title("Proportion of CSAT Scores")
plt.show()
```

```python

bins = [0, 50, 100, 200, 500, 1000, 5000]  
labels = ['0-50', '51-100', '101-200', '201-500', '501-1000', '1001-5000']
df = df.copy()
df['Price_Range'] = pd.cut(df['Item_price'], bins=bins, labels=labels, include_lowest=True)

price_counts = df['Price_Range'].value_counts().sort_index()


plt.figure(figsize=(8,5))
price_counts.plot(kind='bar', color='skyblue')
plt.title("Most Common Price Ranges for Items")
plt.xlabel("Price Range")
plt.ylabel("Number of Items")
plt.show()
```
