```python
trainer_df = pd.DataFrame({
    'Date':['2020/12/01','2020/12/02','2020/12/03'],
    'Trainer': ['Alex','Jordan','Taylor']})
```

```python
trainer_df['Date'] = pd.to_datetime(trainer_df['Date'])
 
# Now merge
merged_df = pd.merge(df, trainer_df, on='Date', how='left')
 
print(merged_df)
```

```python
## Group by Duration and Calculate Average Calories

### Groups the dataset by the Duration column (i.e., how long each activity lasted).

### For each unique Duration, it computes the average Calories burned.

### Returns a Series with Duration as the index and average Calories as values.
 
 
# Group by Duration and get average Calories

avg_calories_by_duration = df.groupby('Duration')['Calories'].mean()
 
# Group by Pulse range (bucketed)

df['PulseRange'] = pd.cut(df['Pulse'], bins=[100, 110, 120, 130, 140, 150, 180])

grouped_stats = df.groupby('PulseRange', observed=False).agg({

    'Calories': 'mean',

    'Maxpulse': 'max'

})
 
print(grouped_stats)
 
```
