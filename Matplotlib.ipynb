```python
import matplotlib.pyplot as plt
```

```python
m = ['Jan','Feb','Mar','Apr','May']
s = [200,100,400,300,400]
plt.plot(m,s,color="blue",linestyle="-.",marker=".")
plt.title("Month vs Sales")
plt.xlabel("Month")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
team = ["India","Australia","New Zealand","England","Sri Lanka","West Indies"]
high_scores = [420,438,391,498,430,411]
plt.bar(team,high_scores,color = "blue")
plt.title("Team vs Highest Score")
plt.xlabel("Team")
plt.ylabel("Highest scores")
plt.grid(False)
plt.show()
```

```python
jer = [18,7,45,10,12]
hi = [83,30,49,100,35]
plt.scatter(jer,hi,color = "black")
plt.title("Jersey no vs No.of centuries")
plt.xlabel("Jersey")
plt.ylabel("Centuries")
plt.grid(True)
plt.show()
```

```python
medals = [20,100,80,30,30,10]
coun = ["India","China","USA","Japan","Brazil","Pakistan"]
colors = ["blue","red","yellow","pink","green","lightgreen"]
plt.pie(medals,labels=coun,colors=colors,autopct="%1.1f%%",startangle=90)
plt.title("Country and its olympic medals")
plt.show()
```

```python
import numpy as np
ages = np.random.randint(18,72,75)
plt.hist(ages,bins=10,color = "grey",edgecolor="white")
plt.title("Age distribution")
plt.xlabel("Ages")
plt.ylabel("Count")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y1 = [1,4,9,16,25,36]
y2 = [1,2,6,12,20,30]
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y1,'-.',marker="*")
plt.title("Line plot")
plt.subplot(1,2,2)
plt.bar(x,y2,color = "yellow")
plt.title("Bar plot")
plt.tight_layout()
```

```python
plt.style.use('ggplot')
x = [1,2,3,4,5]
sales = [10,20,30,40,50]
profits = [3,12,9,89,12]
plt.plot(x,sales,label = "Sales",marker = "*")
plt.plot(x,profits,label = "Profits",marker = ".")
plt.title("Sales vs Profit Trend")
plt.xlabel("Quarter")
plt.ylabel("Amount")
plt.legend()
plt.savefig("save_this_plot.jpg")
plt.show()
```

```python
avg_pulse_by_duration = df.groupby("Duration")["Pulse"].mean().reset_index()
```

```python
avg_pulse_by_duration
```

```python
plt.figure(figsize=(10,5))
plt.bar(avg_pulse_by_duration["Duration"],avg_pulse_by_duration["Pulse"],color="black")
plt.xlabel("Duration")
plt.ylabel("Avg_pulse")
plt.savefig("save3.jpg")
plt.show()
```

```python
plt.scatter(df["Duration"],df["Calories"],color = "black")
plt.title("Duration vs Calories")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.savefig("scatterplot.jpg")
plt.grid(True)
```

```python
plt.hist(df["Calories"],bins = 20, color = "yellow",edgecolor = "white")
plt.title("Histogram")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.savefig("Histogram.jpg")
plt.show()
```

```python

labels = ["<=120","121-140","141-160",">160"]
bins = [0,120,140,160,df["Maxpulse"].max()+10]
df['Maxpulse_Range'] = pd.cut(
    df['Maxpulse'],
    bins=bins,
    labels=labels,
    right=True, 
    include_lowest=True
)
range_count = df["Maxpulse_Range"].value_counts().sort_index()

plt.pie(range_count,labels = range_count.index,autopct="%1.1f%%",startangle=90)
plt.title("Maxpulse distribution")
plt.savefig("piechart.jpg")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(df['Date'], df['Pulse'])
axes[0].set_title('Pulse over Date')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Pulse')

axes[1].scatter(df['Duration'], df['Maxpulse'])
axes[1].set_title('Duration vs Maxpulse')
axes[1].set_xlabel('Duration')
axes[1].set_ylabel('Maxpulse')

plt.tight_layout()
plt.savefig("subplot.jpg")
plt.show()
```
